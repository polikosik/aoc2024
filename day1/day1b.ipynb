```python
import pandas as pd
import os

absolute_path = os.getcwd()
relative_path = 'input.txt'
full_path = os.path.join(absolute_path, relative_path)

df = pd.read_csv(full_path, sep='   ', header=None,names=['A','B'])

df
```

```python
slownik = {}

for value in df['A']:
    count = (df['B'] == value).sum()  
    slownik[value] = int(count)

slownik
```

```python
roll = 0

for key in slownik:
    roll = roll + slownik[key] * int(key) 

roll
```
