```python
import pandas as pd
import os

absolute_path = os.getcwd()
relative_path = 'input.txt'
full_path = os.path.join(absolute_path, relative_path)

df = pd.read_csv(full_path, sep='   ', header=None,names=['A','B'])
```

```python
for col in df:
    df[col] = df[col].sort_values(ignore_index=True)

df
```

```python
df['C'] = abs ( df['A'] - df['B'] )

df
```

```python
Total = df['C'].sum()
print(Total)
```
